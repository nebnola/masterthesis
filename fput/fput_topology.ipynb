```python
from sympy import *
from IPython.display import display
```

The Hamiltonian of the FPUT system, where we substituted
$$p_0 = - p_1 - p_2$$
and
$$x_0 = - x_1 - x_2$$
due to the conservation of momentum and center of mass

```python
p1, p2, x1, x2, alpha = symbols('p1 p2 x1 x2 alpha')

H = ((p1+p2)**2 + p1**2 + p2**2 + (x2-x1)**2 + (2*x1+x2)**2 + (2*x2+x1)**2)/2 + alpha/3*((x2-x1)**3 + (2*x1+x2)**3 - (2*x2+x1)**3)
H
```

Solve $\nabla H = 0$

```python
grad = Matrix([diff(H, var) for var in [p1, p2, x1, x2]])
grad
```

```python
eq = Eq(grad, Matrix([0,0,0,0]))
sol = solve(eq, (p1, p2, x1, x2), dict=True)
sol
```

```python
for s in sol:
    print(f"x0={-s[x1]-s[x2]}, x1={s[x1]}, x2={s[x2]}, H={H.subs(s)}")
```

There are four solutions, one with $H=0$ and three with $H=\frac{1}{\alpha^2}$

Let's check the Hessian

```python
hess = Matrix([diff(grad.T, var) for var in [p1, p2, x1, x2]])
hess
```

```python
for s in sol:
    h = hess.subs(s)
    display(h)
    display(Matrix(h.eigenvals(multiple=True)).T.evalf())
```

There is a local minimum at the origin and three saddle points
